# 10. import scikitlearn for calling ml algorithm-linear regression,a supervised learning algorithm 

Linear regression analysis is used to predict the value of a variable based on the value of another variable or collection of variables. 
The variable you want to predict is called the dependent variable.
The variable you are using to predict the other variable's value is called the independent variable.

```python
from sklearn.linear_model import LinearRegression
```

```python
MLA=LinearRegression()
```

# 11.taking 3 columns from train_data which are higly correlated with medal column("athletes", "prev_medals","height") as features(predictors) in train data-independent variables

```python
feature=["athletes", "prev_medals","height"]
features=train_data[feature]
```

# 12.taking the medal column as label,dependent variable in train data

```python
label=train_data["medals"]
```

# 13.fitting both feature and label to the machine learning algorithm (linear regression)- the model has created

```python
MLA.fit(features,label)
```

# 14.now we are putting our testdata to predict number of medals 

```python
test_data_features=test_data[feature]
medals_predicted=MLA.predict(test_data_features) # predict is the atribute of linear regreession
```

```python
medals_predicted # this is the medal prediction of the test data by analysisng the features in test_data(3 columns in test data)
```

```python
medals_predicted.shape
```

# 15. Adding the prediction column to test data

```python
test_data["medals_predicted"]=medals_predicted
```

```python
test_data
```

no.of medals should not be negatives or fractions,to make it perfect .make values less than 0 as 0 and round the fractions

# 16.replace the values in prediction column less than zero to zero 

```python
test_data.loc[test_data["medals_predicted"]<0 ,"medals_predicted"]=0
print(test_data.to_string())
```

# 17.roundoff the fractional values 

```python
test_data["medals_predicted"]=test_data["medals_predicted"].round()
```

```python
print(test_data.to_string())
```

```python
test_data.drop('predictions', inplace=True, axis=1)# droped the column predictions
```
